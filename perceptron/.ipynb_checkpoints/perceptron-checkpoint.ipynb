```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement.csv')
```

```python
print(df.shape)
```

```python
df.head()
```

```python
cgpa = df['cgpa']
placed = df['placed']
resume_score = df['resume_score']
```

```python
plt.plot(cgpa, placed, 'ro')
plt.xlabel('cgpa')
plt.ylabel('placed')
plt.show
```

```python
plt.plot(resume_score, placed, 'ro')
plt.xlabel('resume score')
plt.ylabel('placed')
plt.show
```
